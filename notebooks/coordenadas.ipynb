Importando as bibliotecas:

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.cluster import KMeans
```

Lendo o CSV:

```python
sucupira = pd.read_csv('sucupirafilter.csv',sep=';')
```

Na célula abaixo temos algumas funções úteis para fazer tratamento dos dados:

```python
# Função para gerar listas das notas e das datas sem outliers
def gerarLista(lst):
    
    res = []
    
    for l in lst:
        if l != ' ' and l != '2345.0 4567.0':
            latitude, longitude = l.split(' ')
            row = [float(latitude), float(longitude)]
            res.append(row)
    
    return res
```

Na célula abaixo é realizado a extração das coordenadas das instituições onde ocorrem os cursos e estes dados extraídos são plotados em um gráfico de dispersão:

```python
X = gerarLista(sucupira['(instituicao) coordenadas'].to_list())
X = np.array(X)

plt.xlim([-70, -10])
plt.scatter(X[:, 1], X[:, 0])
```

Agora iremos aplicar o K-Means (K = 2):

```python
kmeans = KMeans(n_clusters=2, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.xlim([-70, -10])
plt.scatter(X[:, 1], X[:, 0], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 1], centroides[:, 0], c='black', s=200, alpha=0.5);
```

Agora iremos aplicar o K-Means (K = 3):

```python
kmeans = KMeans(n_clusters=3, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.xlim([-70, -10])
plt.scatter(X[:, 1], X[:, 0], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 1], centroides[:, 0], c='black', s=200, alpha=0.5);
```

```python
kmeans = KMeans(n_clusters=4, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.xlim([-70, -10])
plt.scatter(X[:, 1], X[:, 0], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 1], centroides[:, 0], c='black', s=200, alpha=0.5);
```
