Importando as bibliotecas:

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.cluster import KMeans
```

Lendo o CSV:

```python
sucupira = pd.read_csv('sucupirafilter.csv',sep=';')
```

Na célula abaixo temos algumas funções úteis para fazer tratamento dos dados:

```python
# Função para extrair o Ano da data e converte-lo de String para Int
def extrairAno(dt):
    _, _, ano = dt.split('/')
    return int(ano)

# Função para converter a nota de String para Int
def converterNota(nota):
    return int(nota)

# Função para gerar listas das notas e das datas sem outliers
def gerarLista(a, b):
    
    res = []
    
    for i in range(0, len(a)):
        if a[i] != 'A' and b[i] != '-':
            row = [extrairAno(b[i]), converterNota(a[i])]
            res.append(row)
    
    return res
```

Na célula abaixo é realizado a extração das Notas e das Datas de Início dos cursos
e estes dados extraídos são plotados em um gráfico de dispersão:

```python
X = gerarLista(sucupira['(curso) nota_curso'].to_list(), sucupira['(curso) data_inicio'].to_list())
X = np.array(X)

plt.scatter(X[:, 0], X[:, 1])
```

Agora iremos aplicar o K-Means (K = 2):

```python
kmeans = KMeans(n_clusters=2, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', s=200, alpha=0.5);
```

Agora iremos aplicar o K-Means (K = 3):

```python
kmeans = KMeans(n_clusters=3, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', s=200, alpha=0.5);
```

```python
kmeans = KMeans(n_clusters=4, n_init=1)
kmeans.fit(X)
Y = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='viridis')

centroides = kmeans.cluster_centers_
plt.scatter(centroides[:, 0], centroides[:, 1], c='black', s=200, alpha=0.5);
```
